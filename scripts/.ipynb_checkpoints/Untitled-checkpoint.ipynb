```python
import numpy as np
import cv2
import rosbag
import open3d as o3d
import plotly.graph_objects as go
import plotly.express as px
from pathlib import Path
from scipy.spatial.transform import Rotation

# ── LiDAR → IMU extrinsic (from FAST-LIO2 config, pixracer IMU) ──────────────
# These define the LiDAR's pose IN the IMU frame
# i.e. T_imu_lidar

t_imu_lidar = np.array([0.015187, 0.027072, 0.057305])

R_imu_lidar = np.array([
    [-0.944173,  0.016684, -0.329028],
    [-0.021802, -0.999692,  0.011871],
    [-0.328728,  0.018381,  0.944246]
])

# Build 4x4 homogeneous transform T_imu_lidar
T_imu_lidar = np.eye(4)
T_imu_lidar[:3, :3] = R_imu_lidar
T_imu_lidar[:3,  3] = t_imu_lidar

print("T_imu_lidar (LiDAR → IMU):")
print(np.round(T_imu_lidar, 4))

# ── IMU → Camera extrinsic (from Kalibr, cam0 = infra1) ──────────────────────
# T_cam_imu is given directly as a 4x4 matrix

T_cam_imu = np.array([
    [ 0.9999291120311481,  0.010346410806480567, -0.005892596713035742,  0.03395327954001527],
    [-0.010472912576459811, 0.9997062357820703,  -0.021857727251679743, -0.01771990936265375],
    [ 0.00566471665352888,  0.02191789045201512,   0.9997437256934238,  -0.008440895188318377],
    [ 0.0,                  0.0,                   0.0,                  1.0                 ]
])

print("\nT_cam_imu (IMU → Camera):")
print(np.round(T_cam_imu, 4))

# ── Compose: T_cam_lidar = T_cam_imu @ T_imu_lidar ───────────────────────────
T_cam_lidar = T_cam_imu @ T_imu_lidar

R_cam_lidar = T_cam_lidar[:3, :3]
t_cam_lidar = T_cam_lidar[:3,  3]

print("\nT_cam_lidar (LiDAR → Camera, composed):")
print(np.round(T_cam_lidar, 4))
print(f"\nTranslation (meters): {np.round(t_cam_lidar, 4)}")

# ── Camera intrinsics (from Kalibr cam0) ─────────────────────────────────────
# intrinsics: [fx, fy, cx, cy]
fx, fy, cx, cy = 436.65111454554625, 435.15726627794936, \
                 426.3181418502217,  240.606365806057

K = np.array([
    [fx,  0,  cx],
    [ 0, fy,  cy],
    [ 0,  0,   1]
])

# Distortion coefficients [k1, k2, p1, p2] (radtan model)
# Note: image_rect_raw is already rectified so distortion is removed
# Set to zero for rectified images
D = np.zeros(4)

image_width  = 848
image_height = 480

print("\nK (camera intrinsics):")
print(np.round(K, 2))
print(f"\nImage resolution: {image_width} x {image_height}")
print(f"\nNote: using /d455/infra1/image_rect_raw — already rectified, D=0")
```

```python
def extract_synced_frames(bag_path, max_frames=100, time_threshold=0.05):
    """
    Extract synchronized LiDAR scan + camera image pairs from bag.
    
    Matches each LiDAR scan to the nearest camera frame within
    time_threshold seconds.
    
    Returns list of (timestamp, points, image) tuples.
    """
    bag_path = str(bag_path)
    
    lidar_msgs  = []
    camera_msgs = []

    print("Reading bag...")
    with rosbag.Bag(bag_path) as bag:
        # Extract LiDAR messages
        for topic, msg, t in bag.read_messages('/livox/lidar'):
            ts = msg.header.stamp.secs + msg.header.stamp.nsecs * 1e-9
            lidar_msgs.append((ts, msg))

        # Extract camera messages
        for topic, msg, t in bag.read_messages('/d455/infra1/image_rect_raw'):
            ts = msg.header.stamp.secs + msg.header.stamp.nsecs * 1e-9
            camera_msgs.append((ts, msg))

    print(f"LiDAR messages:  {len(lidar_msgs)}")
    print(f"Camera messages: {len(camera_msgs)}")

    if not lidar_msgs or not camera_msgs:
        print("ERROR: No messages found — check topic names")
        return []

    # Build camera timestamp array for fast nearest-neighbor lookup
    cam_times = np.array([t for t, _ in camera_msgs])

    synced = []
    for lidar_ts, lidar_msg in lidar_msgs[:max_frames]:
        # Find nearest camera frame
        diffs   = np.abs(cam_times - lidar_ts)
        best    = np.argmin(diffs)
        best_dt = diffs[best]

        if best_dt > time_threshold:
            continue   # no close enough camera frame

        cam_msg = camera_msgs[best][1]

        # Decode LiDAR points from Livox CustomMsg
        pts = decode_livox_msg(lidar_msg)

        # Decode camera image
        img = decode_image_msg(cam_msg)

        if pts is not None and img is not None:
            synced.append((lidar_ts, pts, img, best_dt))

    print(f"\nSynced pairs found: {len(synced)}")
    if synced:
        dts = [s[3] for s in synced]
        print(f"Time offset stats: "
              f"mean={np.mean(dts)*1000:.1f}ms  "
              f"max={np.max(dts)*1000:.1f}ms")

    return synced


def decode_livox_msg(msg):
    """
    Decode livox_ros_driver2/CustomMsg to Nx3 numpy array.
    """
    try:
        pts = np.array([
            [p.x, p.y, p.z]
            for p in msg.points
            if not (p.x == 0 and p.y == 0 and p.z == 0)
        ], dtype=np.float32)
        return pts if len(pts) > 0 else None
    except Exception as e:
        print(f"LiDAR decode error: {e}")
        return None


def decode_image_msg(msg):
    """
    Decode sensor_msgs/Image to HxWx1 numpy array.
    Handles mono8 and mono16 encodings.
    """
    try:
        if msg.encoding == 'mono8':
            img = np.frombuffer(msg.data, dtype=np.uint8).reshape(
                msg.height, msg.width
            )
        elif msg.encoding == 'mono16':
            img = np.frombuffer(msg.data, dtype=np.uint16).reshape(
                msg.height, msg.width
            )
            img = (img / 256).astype(np.uint8)
        else:
            img = np.frombuffer(msg.data, dtype=np.uint8).reshape(
                msg.height, msg.width, -1
            )

        # Convert grayscale to 3-channel so projection code works uniformly
        if img.ndim == 2:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

        return img
    except Exception as e:
        print(f"Image decode error: {e}")
        return None


# Run extraction
BAG_PATH = Path.home() / 'catkin_ws/data/camtest.bag'

synced_frames = extract_synced_frames(
    BAG_PATH,
    max_frames=200,
    time_threshold=0.05   # 50ms max time gap
)
```

```python
def project_lidar_to_image_uav(points, K, R_cam_lidar, t_cam_lidar,
                                 image, min_depth=0, max_depth=100.0):
    """
    Project LiDAR points onto UAV camera image.
    
    Differences from KITTI version:
    - No R_rect (single camera, already rectified)
    - Shorter depth range (indoor UAV ~20m max)
    - R_cam_lidar and t_cam_lidar are the composed transform
    
    Returns:
        pts_3d:   Nx3  LiDAR XYZ for points inside image
        pts_2d:   Nx2  pixel (u, v) coordinates
        colors:   Nx3  RGB sampled from image
        depths:   N    distance from camera in meters
    """
     H, W = image.shape[:2]
    xyz  = points[:, :3]

    # Step 1 — Transform LiDAR points to camera frame
    pts_cam = (R_lidar2cam @ xyz.T).T + t_lidar2cam   # Nx3

    # Apply rectification rotation
    pts_rect = (R_rect @ pts_cam.T).T                  # Nx3

    # Step 2 — Remove points behind the camera (z <= 0)
    front_mask = pts_rect[:, 2] > 0
    pts_rect   = pts_rect[front_mask]
    xyz_front  = xyz[front_mask]

    # Step 3 — Project to image plane
    Z = pts_rect[:, 2]
    u = (K[0, 0] * pts_rect[:, 0] / Z + K[0, 2]).astype(int)
    v = (K[1, 1] * pts_rect[:, 1] / Z + K[1, 2]).astype(int)

    # Step 4 — Keep only points that project inside the image
    in_bounds = (u >= 0) & (u < W) & (v >= 0) & (v < H)
    u         = u[in_bounds]
    v         = v[in_bounds]
    Z         = Z[in_bounds]
    xyz_valid = xyz_front[in_bounds]

    # Step 5 — Sample color at each pixel
    colors = image[v, u]   # Nx3 RGB

    return xyz_valid, np.stack([u, v], axis=1), colors, Z



# Test on first synced frame
if synced_frames:
    ts, pts, img, dt = synced_frames[0]

    pts_3d, pts_2d, colors, depths = project_lidar_to_image_uav(
        pts, K, R_cam_lidar, t_cam_lidar, img
    )

    print(f"Timestamp:            {ts:.3f}")
    print(f"LiDAR-camera offset:  {dt*1000:.1f}ms")
    print(f"Raw LiDAR points:     {len(pts):,}")
    print(f"Projected into image: {len(pts_3d):,} "
          f"({100*len(pts_3d)/max(len(pts),1):.1f}%)")

    if len(pts_3d) > 0:
        print(f"\nDepth range: {depths.min():.2f}m – {depths.max():.2f}m")
        print(f"XYZ range:")
        print(f"  X: {pts_3d[:,0].min():.2f} to {pts_3d[:,0].max():.2f}")
        print(f"  Y: {pts_3d[:,1].min():.2f} to {pts_3d[:,1].max():.2f}")
        print(f"  Z: {pts_3d[:,2].min():.2f} to {pts_3d[:,2].max():.2f}")
```
