```python
import numpy as np
import open3d as o3d
import rosbag
import copy
import json
from pathlib import Path
from sensor_msgs.point_cloud2 import read_points

print(f"Open3D  {o3d.__version__}")
print(f"NumPy   {np.__version__}")
```

```python
# ---- Bag files ---------------------------------------------------------------
# First entry = reference. All other sessions align into its coordinate frame.
BAG_FILES = [
    Path("~/catkin_ws/data/Apr4_Palio2.bag").expanduser(),    # reference (small)
    Path("~/catkin_ws/data/Apr4_droneparkscan.bag").expanduser(),   # target   (large 4.4GB)
]

# How many frames to skip between reads. 1 = every frame (slow, small bags).
# 10 = every 10th frame (10x faster, large bags). Same map quality for alignment.
# Set per bag — reference reads everything, big bag skips most frames.
EVERY_NTH = [1, 10]

LIDAR_TOPIC = "/cloud_registered"   # use /livox/lidar if FAST-LIO2 wasn't running

# ---- Output ------------------------------------------------------------------
OUTPUT_DIR = Path("~/maps").expanduser()
OUTPUT_MAP = OUTPUT_DIR / "comprehensive.pcd"
OUTPUT_LOG = OUTPUT_DIR / "alignment_log.json"

# ---- Voxel sizes -------------------------------------------------------------
# ICP_VOXEL    : downsampled cloud used for ICP computation only (not output)
# RANSAC_VOXEL : downsampled cloud used for FPFH + RANSAC (coarser = faster)
# VOXEL_SIZE   : final output deduplication voxel
ICP_VOXEL    = 0.10
RANSAC_VOXEL = 0.15
VOXEL_SIZE   = 0.05

# ---- Ground extraction -------------------------------------------------------
GROUND_DIST_THRESHOLD = 0.15   # metres — RANSAC inlier threshold for plane fit
GROUND_REMOVE_Z       = 0.30   # metres — strip points below this after levelling
                                # removes ground from FPFH input

# ---- RANSAC ------------------------------------------------------------------
RANSAC_DIST    = RANSAC_VOXEL * 2.0   # looser than indoors — outdoor features larger
NORMAL_RADIUS_R  = RANSAC_VOXEL * 2
FEATURE_RADIUS_R = RANSAC_VOXEL * 5

# ---- ICP ---------------------------------------------------------------------
# Outdoor thresholds are looser than indoor — vegetation and sparse density
# mean true correspondences can be further apart than indoor wall points
ICP_DIST_COARSE = 2.0    # metres
ICP_DIST_FINE   = 0.3    # metres
ICP_MAX_ITER    = 200
NORMAL_RADIUS   = ICP_VOXEL * 2

# ---- Merge -------------------------------------------------------------------
MIN_FITNESS = 0.15   # outdoor fitness is naturally lower than indoor

# ---- Session colours ---------------------------------------------------------
SESSION_COLORS = [
    [0.2, 0.6, 1.0],   # blue   — reference
    [1.0, 0.4, 0.2],   # orange — target 1
    [0.2, 0.9, 0.4],   # green  — target 2
    [0.9, 0.2, 0.9],   # purple — target 3
]

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("Config loaded.")
for i, (p, nth) in enumerate(zip(BAG_FILES, EVERY_NTH)):
    label = "REFERENCE" if i == 0 else f"TARGET   {i}"
    print(f"  {label} : {p.name}  (every {nth}th frame)")
print(f"\n  Topic         : {LIDAR_TOPIC}")
print(f"  RANSAC voxel  : {RANSAC_VOXEL} m")
print(f"  ICP voxel     : {ICP_VOXEL} m")
print(f"  Output voxel  : {VOXEL_SIZE} m")
print(f"  ICP coarse    : {ICP_DIST_COARSE} m")
print(f"  ICP fine      : {ICP_DIST_FINE} m")
```

```python
def inspect_bag(path):
    print(f"\n[{path.name}]")
    with rosbag.Bag(str(path)) as bag:
        info     = bag.get_type_and_topic_info()
        duration = bag.get_end_time() - bag.get_start_time()
        print(f"  Duration : {duration:.1f} s")
        for topic, tinfo in sorted(info.topics.items()):
            marker = "  <-- will read" if topic == LIDAR_TOPIC else ""
            print(f"  {tinfo.message_count:>6} msgs  {topic}{marker}")

for bag_path in BAG_FILES:
    inspect_bag(bag_path)
```

```python
def extract_pointcloud2(msg):
    """Decode sensor_msgs/PointCloud2 -> (N, 3) float32."""
    pts = np.array(
        list(read_points(msg, field_names=("x", "y", "z"), skip_nans=True)),
        dtype=np.float32
    )
    return pts if len(pts) > 0 else None


def extract_livox_custommsg(msg):
    """Decode livox_ros_driver2/CustomMsg -> (N, 3) float32."""
    pts = np.array(
        [[p.x, p.y, p.z] for p in msg.points
         if not (p.x == 0.0 and p.y == 0.0 and p.z == 0.0)],
        dtype=np.float32
    )
    return pts if len(pts) > 0 else None


def decode_msg(msg, topic):
    if topic == "/livox/lidar":
        return extract_livox_custommsg(msg)
    return extract_pointcloud2(msg)


print("Extraction functions ready.")
```

```python
def accumulate_from_bag(bag_path, topic, every_nth=1):
    all_points = []
    frame_idx  = 0
    kept       = 0

    with rosbag.Bag(str(bag_path)) as bag:
        total     = bag.get_message_count(topic_filters=[topic])
        effective = max(1, total // every_nth)
        print(f"  {total} total frames, keeping every {every_nth}th (~{effective})")

        for _, msg, _ in bag.read_messages(topics=[topic]):
            frame_idx += 1
            if frame_idx % every_nth != 0:
                continue
            pts = decode_msg(msg, topic)
            if pts is None or len(pts) == 0:
                continue
            all_points.append(pts)
            kept += 1
            if kept % 50 == 0:
                print(f"    {kept}/{effective} frames  "
                      f"({sum(len(p) for p in all_points):,} pts so far)")

    if not all_points:
        raise RuntimeError(
            f"No points from {bag_path.name} on {topic}.\n"
            f"Run Cell 3 to confirm the topic exists."
        )

    stacked = np.vstack(all_points)
    stacked = stacked[~np.all(stacked == 0, axis=1)]
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(stacked)
    return pcd, kept


raw_clouds = []
for i, bag_path in enumerate(BAG_FILES):
    nth   = EVERY_NTH[i]
    label = "REFERENCE" if i == 0 else f"TARGET {i}"
    print(f"\nExtracting {label}: {bag_path.name} ...")
    pcd, n = accumulate_from_bag(bag_path, LIDAR_TOPIC, every_nth=nth)
    raw_clouds.append(pcd)
    print(f"  Done: {n} frames -> {len(pcd.points):,} points")

reference = raw_clouds[0]
targets   = raw_clouds[1:]
print(f"\nExtracted {len(raw_clouds)} sessions.")
```

```python
def inspect_cloud(pcd, label):
    bb  = pcd.get_axis_aligned_bounding_box()
    ctr = bb.get_center()
    ext = bb.get_extent()
    print(f"[{label}]")
    print(f"  Points : {len(pcd.points):,}")
    print(f"  Center : ({ctr[0]:.2f}, {ctr[1]:.2f}, {ctr[2]:.2f})")
    print(f"  X      : {bb.min_bound[0]:.2f} -> {bb.max_bound[0]:.2f}  ({ext[0]:.2f} m)")
    print(f"  Y      : {bb.min_bound[1]:.2f} -> {bb.max_bound[1]:.2f}  ({ext[1]:.2f} m)")
    print(f"  Z      : {bb.min_bound[2]:.2f} -> {bb.max_bound[2]:.2f}  ({ext[2]:.2f} m)")
    print()
    return ctr

centers = []
ref_ctr = inspect_cloud(reference, f"REFERENCE  ({BAG_FILES[0].name})")
centers.append(ref_ctr)
for i, tgt in enumerate(targets):
    ctr = inspect_cloud(tgt, f"TARGET {i+1}  ({BAG_FILES[i+1].name})")
    centers.append(ctr)

print("Frame offsets from reference:")
for i in range(1, len(centers)):
    offset = centers[0] - centers[i]
    dist   = np.linalg.norm(offset)
    print(f"  Target {i}: dX={offset[0]:.1f}  dY={offset[1]:.1f}  dZ={offset[2]:.1f}  "
          f"total={dist:.1f} m")
    if dist > 5.0:
        print(f"    -> Large offset ({dist:.0f} m). Direct RANSAC would fail.")
        print(f"       Ground-plane levelling + XY normalisation will handle this.")
```

```python
print(f"Building {ICP_VOXEL}m downsampled clouds for ICP ...")
print(f"(Full clouds used for final merge — downsampling only speeds up alignment)\n")

ref_icp = reference.voxel_down_sample(ICP_VOXEL)
ref_icp.estimate_normals(
    o3d.geometry.KDTreeSearchParamHybrid(radius=NORMAL_RADIUS, max_nn=30)
)
print(f"  Reference : {len(reference.points):,} -> {len(ref_icp.points):,} pts")

targets_icp = []
for i, tgt in enumerate(targets):
    t = tgt.voxel_down_sample(ICP_VOXEL)
    t.estimate_normals(
        o3d.geometry.KDTreeSearchParamHybrid(radius=NORMAL_RADIUS, max_nn=30)
    )
    targets_icp.append(t)
    print(f"  Target {i+1}  : {len(tgt.points):,} -> {len(t.points):,} pts")
```

```python
def extract_ground_and_level(pcd, dist_threshold=0.15, label=""):
    """
    Fit ground plane, return (plane_model, inlier_indices, T_level).
    T_level is the 4x4 transform that rotates the cloud upright and
    translates it so the ground sits at Z=0.
    """
    plane_model, inliers = pcd.segment_plane(
        distance_threshold=dist_threshold,
        ransac_n=3,
        num_iterations=1000
    )
    [a, b, c, d] = plane_model
    normal = np.array([a, b, c])
    normal /= np.linalg.norm(normal)
    if normal[2] < 0:
        normal = -normal   # ensure normal points upward

    if label:
        print(f"  [{label}] plane: {a:.3f}x + {b:.3f}y + {c:.3f}z + {d:.3f} = 0")
        print(f"  [{label}] inliers: {len(inliers):,} / {len(pcd.points):,} pts  "
              f"({100*len(inliers)/len(pcd.points):.1f}%)")

    # Rotation to align ground normal with +Z
    z_axis = np.array([0.0, 0.0, 1.0])
    axis   = np.cross(normal, z_axis)
    angle  = np.arccos(np.clip(np.dot(normal, z_axis), -1.0, 1.0))

    if np.linalg.norm(axis) < 1e-6:
        R = np.eye(3)
    else:
        axis /= np.linalg.norm(axis)
        K = np.array([
            [ 0,       -axis[2],  axis[1]],
            [ axis[2],  0,       -axis[0]],
            [-axis[1],  axis[0],  0      ]
        ])
        R = np.eye(3) + np.sin(angle) * K + (1 - np.cos(angle)) * (K @ K)

    T_level = np.eye(4)
    T_level[:3, :3] = R

    # Find ground Z after rotation and translate to Z=0
    tmp = copy.deepcopy(pcd)
    tmp.transform(T_level)
    ground_z = np.percentile(np.asarray(tmp.points)[:, 2], 2)
    T_level[2, 3] = -ground_z

    return plane_model, inliers, T_level


print("Extracting ground planes and levelling ...")

print("\nReference:")
_, ref_inliers, T_ref_level = extract_ground_and_level(
    ref_icp, GROUND_DIST_THRESHOLD, label="REF"
)

print("\nTargets:")
T_tgt_levels = []
for i, tgt in enumerate(targets_icp):
    _, tgt_inliers, T_tgt_level = extract_ground_and_level(
        tgt, GROUND_DIST_THRESHOLD, label=f"TGT {i+1}"
    )
    T_tgt_levels.append(T_tgt_level)

# Apply levelling
ref_levelled = copy.deepcopy(ref_icp)
ref_levelled.transform(T_ref_level)

targets_levelled = []
for tgt, T in zip(targets_icp, T_tgt_levels):
    t = copy.deepcopy(tgt)
    t.transform(T)
    targets_levelled.append(t)

print("\nZ ranges after levelling (ground should be near 0):")
ref_z = np.asarray(ref_levelled.points)[:, 2]
print(f"  Reference : {ref_z.min():.2f} -> {ref_z.max():.2f} m")
for i, tgt in enumerate(targets_levelled):
    tgt_z = np.asarray(tgt.points)[:, 2]
    print(f"  Target {i+1}  : {tgt_z.min():.2f} -> {tgt_z.max():.2f} m")
```

```python
def remove_ground_points(pcd, z_threshold):
    pts  = np.asarray(pcd.points)
    mask = pts[:, 2] > z_threshold
    above = o3d.geometry.PointCloud()
    above.points = o3d.utility.Vector3dVector(pts[mask])
    if pcd.has_normals():
        norms = np.asarray(pcd.normals)
        above.normals = o3d.utility.Vector3dVector(norms[mask])  # same type as points
    if pcd.has_colors():
        cols = np.asarray(pcd.colors)
        above.colors = o3d.utility.Vector3dVector(cols[mask])
    return above


def normalise_xy(pcd):
    ctr    = pcd.get_axis_aligned_bounding_box().get_center()
    T      = np.eye(4)
    T[0,3] = -ctr[0]
    T[1,3] = -ctr[1]
    norm   = copy.deepcopy(pcd)
    norm.transform(T)
    return norm, ctr


def preprocess_for_ransac(pcd, voxel, normal_r, feature_r):
    down = pcd.voxel_down_sample(voxel)
    down.estimate_normals(
        o3d.geometry.KDTreeSearchParamHybrid(radius=normal_r, max_nn=30)
    )
    fpfh = o3d.pipelines.registration.compute_fpfh_feature(
        down,
        o3d.geometry.KDTreeSearchParamHybrid(radius=feature_r, max_nn=100)
    )
    return down, fpfh


def make_T(tx=0, ty=0, tz=0):
    T = np.eye(4)
    T[0,3], T[1,3], T[2,3] = tx, ty, tz
    return T


ransac_results  = []
final_transforms = []

for i, (tgt_lev, T_tgt_lev) in enumerate(zip(targets_levelled, T_tgt_levels)):

    print(f"\n--- Target {i+1}: {BAG_FILES[i+1].name} ---")

    # Step 1: remove ground
    ref_above = remove_ground_points(ref_levelled, GROUND_REMOVE_Z)
    tgt_above = remove_ground_points(tgt_lev,      GROUND_REMOVE_Z)
    print(f"  Above-ground pts — ref: {len(ref_above.points):,}  "
          f"tgt: {len(tgt_above.points):,}")

    # Step 2: normalise XY
    ref_norm, ref_xy_ctr = normalise_xy(ref_above)
    tgt_norm, tgt_xy_ctr = normalise_xy(tgt_above)
    print(f"  XY origin removed — ref: ({ref_xy_ctr[0]:.1f},{ref_xy_ctr[1]:.1f})  "
          f"tgt: ({tgt_xy_ctr[0]:.1f},{tgt_xy_ctr[1]:.1f})")

    # Step 3: FPFH features
    print(f"  Computing FPFH features (voxel={RANSAC_VOXEL}m) ...")
    ref_down, ref_fpfh = preprocess_for_ransac(
        ref_norm, RANSAC_VOXEL, NORMAL_RADIUS_R, FEATURE_RADIUS_R)
    tgt_down, tgt_fpfh = preprocess_for_ransac(
        tgt_norm, RANSAC_VOXEL, NORMAL_RADIUS_R, FEATURE_RADIUS_R)
    print(f"  RANSAC pts — ref: {len(ref_down.points):,}  tgt: {len(tgt_down.points):,}")

    # Step 4: RANSAC
    print(f"  Running RANSAC (dist={RANSAC_DIST:.2f}m) ...")
    r = o3d.pipelines.registration.registration_ransac_based_on_feature_matching(
        tgt_down, ref_down, tgt_fpfh, ref_fpfh,
        mutual_filter=True,
        max_correspondence_distance=RANSAC_DIST,
        estimation_method=o3d.pipelines.registration.TransformationEstimationPointToPoint(False),
        ransac_n=4,
        checkers=[
            o3d.pipelines.registration.CorrespondenceCheckerBasedOnEdgeLength(0.9),
            o3d.pipelines.registration.CorrespondenceCheckerBasedOnDistance(RANSAC_DIST),
        ],
        criteria=o3d.pipelines.registration.RANSACConvergenceCriteria(4_000_000, 500)
    )
    ransac_results.append(r)
    print(f"  RANSAC fitness : {r.fitness:.4f}  RMSE : {r.inlier_rmse:.4f} m")

    if r.fitness == 0:
        print("  WARNING: fitness=0. Try lowering RANSAC_VOXEL to 0.15 in Cell 2.")

    # Compose full transform:
    #   level target -> remove XY offset -> RANSAC rotation ->
    #   restore ref XY offset -> unlevel back to ref frame
    T_tgt_to_norm = make_T(-tgt_xy_ctr[0], -tgt_xy_ctr[1])
    T_norm_to_ref = make_T( ref_xy_ctr[0],  ref_xy_ctr[1])
    T_ransac_levelled = T_norm_to_ref @ r.transformation @ T_tgt_to_norm
    T_full = np.linalg.inv(T_ref_level) @ T_ransac_levelled @ T_tgt_lev

    final_transforms.append(T_full)
    print(f"  Full transform translation: "
          f"({T_full[0,3]:.2f}, {T_full[1,3]:.2f}, {T_full[2,3]:.2f})")
```

```python
SESSION_TO_CHECK = 0

tgt = targets_icp[SESSION_TO_CHECK]
T   = final_transforms[SESSION_TO_CHECK]
r   = ransac_results[SESSION_TO_CHECK]

ref_vis = copy.deepcopy(ref_icp)
tgt_vis = copy.deepcopy(tgt)

ref_vis.paint_uniform_color([1.0, 1.0, 0.0])   # yellow
tgt_vis.transform(T)
tgt_vis.paint_uniform_color([0.0, 0.9, 0.9])   # cyan

print(f"Target {SESSION_TO_CHECK+1}: {BAG_FILES[SESSION_TO_CHECK+1].name}")
print(f"  RANSAC fitness : {r.fitness:.4f}")
print(f"  RANSAC RMSE    : {r.inlier_rmse*100:.2f} cm")
print("\n  Yellow = reference  |  Cyan = target after RANSAC")
print("  Structures should roughly overlap. Close the window to continue.")

o3d.visualization.draw_geometries([ref_vis, tgt_vis],
                                  window_name=f"After RANSAC — Target {SESSION_TO_CHECK+1}",
                                  width=1280, height=720)
```

```python
def icp_pass(src, tgt, initial_transform, dist_threshold, max_iter, label=""):
    result = o3d.pipelines.registration.registration_icp(
        src, tgt,
        dist_threshold,
        initial_transform,
        o3d.pipelines.registration.TransformationEstimationPointToPlane(),
        o3d.pipelines.registration.ICPConvergenceCriteria(max_iteration=max_iter)
    )
    if label:
        print(f"  {label:<24} fitness={result.fitness:.4f}  "
              f"RMSE={result.inlier_rmse*100:.2f} cm")
    return result


# Re-estimate normals on levelled ICP clouds for Point-to-Plane
print("Estimating normals on levelled ICP clouds ...")
ref_levelled.estimate_normals(
    o3d.geometry.KDTreeSearchParamHybrid(radius=NORMAL_RADIUS, max_nn=30)
)
targets_levelled_normals = []
for tgt in targets_levelled:
    t = copy.deepcopy(tgt)
    t.estimate_normals(
        o3d.geometry.KDTreeSearchParamHybrid(radius=NORMAL_RADIUS, max_nn=30)
    )
    targets_levelled_normals.append(t)

icp_results = []
icp_transforms = []

for i, (tgt_lev, T_ransac, T_tgt_lev) in enumerate(
        zip(targets_levelled_normals, final_transforms, T_tgt_levels)):

    print(f"\nICP — Target {i+1} ({BAG_FILES[i+1].name})")

    r1 = icp_pass(tgt_lev, ref_levelled, T_ransac,
                  ICP_DIST_COARSE, ICP_MAX_ITER, "Pass 1 (coarse 2.0m)")
    r2 = icp_pass(tgt_lev, ref_levelled, r1.transformation,
                  ICP_DIST_FINE, ICP_MAX_ITER, "Pass 2 (fine   0.3m)")

    icp_results.append(r2)

    # Compose back to original (non-levelled) coordinate frames
    T_icp_final = np.linalg.inv(T_ref_level) @ r2.transformation @ T_tgt_lev
    icp_transforms.append(T_icp_final)
```

```python
SESSION_TO_CHECK = 0
tgt     = targets_levelled[SESSION_TO_CHECK]
T_r     = final_transforms[SESSION_TO_CHECK]
T_icp   = icp_transforms[SESSION_TO_CHECK]
r_icp   = icp_results[SESSION_TO_CHECK]
r_ransac = ransac_results[SESSION_TO_CHECK]

ref_vis    = copy.deepcopy(ref_icp)
tgt_ransac = copy.deepcopy(tgt)
tgt_fine   = copy.deepcopy(tgt)

ref_vis.paint_uniform_color([1.0, 1.0, 0.0])       # yellow — reference
tgt_ransac.transform(T_r)
tgt_ransac.paint_uniform_color([1.0, 0.5, 0.0])    # orange — after RANSAC
tgt_fine.transform(r_icp.transformation)
tgt_fine.paint_uniform_color([0.0, 0.9, 0.9])      # cyan   — after ICP

print(f"Target {SESSION_TO_CHECK+1}: {BAG_FILES[SESSION_TO_CHECK+1].name}")
print(f"  RANSAC fitness : {r_ransac.fitness:.4f}")
print(f"  ICP    fitness : {r_icp.fitness:.4f}   "
      f"RMSE={r_icp.inlier_rmse*100:.2f} cm")
print("\n  Yellow = reference  |  Orange = after RANSAC  |  Cyan = after ICP")
print("  Close the window to continue.")

o3d.visualization.draw_geometries([ref_vis, tgt_ransac, tgt_fine],
                                  window_name=f"After ICP — Target {SESSION_TO_CHECK+1}",
                                  width=1280, height=720)
```

```python
print(f"{'Session':<35} {'RANSAC fit':>12} {'ICP fit':>10} {'RMSE (cm)':>12} {'Status':>10}")
print("-" * 80)

for i, (r_r, r_i) in enumerate(zip(ransac_results, icp_results)):
    name   = BAG_FILES[i + 1].name
    status = "ACCEPT" if r_i.fitness >= MIN_FITNESS else "REJECT"
    print(f"{name:<35} {r_r.fitness:>10.4f}  {r_i.fitness:>10.4f}  "
          f"{r_i.inlier_rmse*100:>10.2f}  {status:>10}")

print()
print(f"MIN_FITNESS threshold : {MIN_FITNESS}")
print()
print("  > 0.4  excellent  |  0.2-0.4  acceptable  |  < 0.15  rejected")
print("  (outdoor fitness is naturally lower than indoor due to vegetation/sparse density)")
```

```python
def colorize(pcd, color):
    c    = copy.deepcopy(pcd)
    cols = np.tile(color, (len(c.points), 1))
    c.colors = o3d.utility.Vector3dVector(cols)
    return c


merged    = colorize(reference, SESSION_COLORS[0])
merge_log = {}

for i, (r_icp, T_final, tgt) in enumerate(
        zip(icp_results, icp_transforms, targets)):

    name  = BAG_FILES[i + 1].name
    color = SESSION_COLORS[(i + 1) % len(SESSION_COLORS)]

    if r_icp.fitness < MIN_FITNESS:
        print(f"Target {i+1} ({name}): SKIPPED  fitness={r_icp.fitness:.3f} < {MIN_FITNESS}")
        print(f"  If RANSAC looked correct (Cell 13), try lowering RANSAC_VOXEL to 0.15")
        merge_log[name] = {"status": "skipped", "fitness": round(r_icp.fitness, 4)}
        continue

    # Apply transform to FULL resolution original cloud (not ICP downsampled)
    tgt_aligned = colorize(tgt, color)
    tgt_aligned.transform(T_final)
    merged += tgt_aligned

    print(f"Target {i+1} ({name}): merged  ->  {len(merged.points):,} total points")
    merge_log[name] = {
        "status":    "merged",
        "fitness":   round(r_icp.fitness, 4),
        "rmse_m":    round(float(r_icp.inlier_rmse), 5),
        "transform": T_final.tolist()
    }

print(f"\nPre-deduplicate total : {len(merged.points):,} points")
```

```python
merged_clean = merged.voxel_down_sample(VOXEL_SIZE)

before = len(merged.points)
after  = len(merged_clean.points)
print(f"Before dedup : {before:,} points")
print(f"After  dedup : {after:,} points")
print(f"Removed      : {before - after:,}  ({100*(before-after)/before:.1f}%)")
```
